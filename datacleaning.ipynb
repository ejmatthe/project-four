```python
# Imports and dependencies
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
suicide_data = pd.read_csv('resources/master.csv')
```

```python
suicide_data
```

```python
suicide_data_df = suicide_data
suicide_data_df = suicide_data_df.dropna(how='any')
suicide_data_df
```

```python
## Exporting dataframes as CSVs to then use in database loading
suicide_data_df.to_csv("output/suicides.csv", index=False)
```

```python
## Define and run engine for database loading
engine = create_engine(f"sqlite:///data/suicides.sqlite")
Base = automap_base()
Base.prepare(autoload_with=engine)
session = Session(engine)
connection = engine.connect()
```

```python
## Load dataframes into database
suicide_data_df.to_sql("suicides", connection, if_exists="replace")
```
